**Ratings and Reviews**

```python
base_url = base_url.replace("/p/", "/product-reviews/")
base_url = re.sub(r"&sortOrder=.*", "", base_url)  # Remove sorting parameters if already present

# Different sorting options
categories = [
    f"{base_url}&aid=overall&certifiedBuyer=false&sortOrder=MOST_HELPFUL",  # Most Helpful
    f"{base_url}&aid=overall&certifiedBuyer=false&sortOrder=NEGATIVE_FIRST",  # Negative First
    f"{base_url}&aid=overall&certifiedBuyer=false&sortOrder=POSITIVE_FIRST"   # Positive First
]

# Initialize lists
all_ratings = []
all_review_headings = []
all_comments = []

# Set the number of pages to scrape
num_pages = 10  # Adjust based on available pages

# Loop through each category and scrape reviews
for url in categories:
    for i in range(1, num_pages + 1):
        print(f"Scraping page {i}...")

        # Construct URL for each page
        page_url = f"{url}&page={i}"

        try:
            driver.get(page_url)
            sleep(2)  # Allow time for page load

            # Wait for reviews to load
            WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, 'div.XQDdHH.Ga3i8K, div.cDye2S'))
            )

            # Extract Ratings
            page_ratings = [ra.text for ra in driver.find_elements(By.CSS_SELECTOR, 'div.XQDdHH.Ga3i8K, div.cDye2S')]

            # Extract Review Titles
            page_titles = [re.text for re in driver.find_elements(By.CSS_SELECTOR, 'p.z9E0IG')]

            # Extract Review Comments
            page_comments = [co.text for co in driver.find_elements(By.CSS_SELECTOR, 'div.ZmyHeo')]

            # Ensure lists have the same length before adding
            min_len = min(len(page_ratings), len(page_titles), len(page_comments))

            print(f"Ratings found: {len(page_ratings)}, Review Headings: {len(page_titles)}, Comments: {len(page_comments)}")

            # Normalize list sizes (fill missing values with "N/A")
            max_length = max(len(page_ratings), len(page_titles), len(page_comments))
            page_ratings += ["N/A"] * (max_length - len(page_ratings))
            page_titles += ["N/A"] * (max_length - len(page_titles))
            page_comments += ["N/A"] * (max_length - len(page_comments))

            # Append to main lists
            all_ratings.extend(page_ratings)
            all_review_headings.extend(page_titles)
            all_comments.extend(page_comments)

            sleep(2)

        except Exception as e:
            print(f"‚ö†Ô∏è Failed to fetch page {i}: {e}")
            continue  # Skip this page and move to the next one

# Close browser
driver.quit()

# Ensure all lists are of equal length
min_len = min(len(all_ratings), len(all_review_headings), len(all_comments))
df = pd.DataFrame({
    "Rating": all_ratings[:min_len],
    "Title": all_review_headings[:min_len],
    "Review": all_comments[:min_len]
})

# Save to CSV
df.to_csv("flipkart_reviews.csv", index=False)
print("\n‚úÖ Reviews scraped and saved to flipkart_reviews.csv üéâ")
```

**Data Preprocessing**

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_comment(comment):
    comment = str(comment)
    comment = re.sub('<.*?>', '', comment)
    comment = re.sub(r'[^\w\s\']', '', comment)
    comment = comment.lower().strip()
    words = comment.split()
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    return " ".join(words)

data['Title'] = data['Title'].apply(preprocess_comment)
data['Review'] = data['Review'].apply(preprocess_comment)
```

```python
data['Full_Text'] = data['Title'] + " " + data['Review']
```

**Sentiment Analysis**

```python
def classify_sentiment(text):

    analysis = TextBlob(text)

    if analysis.sentiment.polarity > 0.2:
        return "Positive"

    elif analysis.sentiment.polarity < -0.2:
        return "Negative"

    else:
        return "Neutral"

data['Sentiment'] = data['Full_Text'].apply(classify_sentiment)
```

```python
sentiment_counts = data['Sentiment'].value_counts()

plt.figure(figsize=(8, 5))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values)
plt.title("Sentiment Distribution", fontsize=16)
plt.xlabel("Sentiment", fontsize=14)
plt.ylabel("Number of Reviews", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

**WOrd Clouds**

```python
def generate_wordcloud(text, title):
    if text:  # Check if the text is not empty
        wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis("off")
        plt.title(title, fontsize=16)
        plt.show()
    else:
        print(f"No text available for category: {title}. Cannot generate word cloud.")
```

```python
# Generate word clouds for each sentiment category
positive_reviews = " ".join(data[data['Sentiment'] == 'Positive']['Full_Text'])
generate_wordcloud(positive_reviews, "Positive Reviews Word Cloud")
```

**Calculate Sentiment Score-Based comments**

```python
def calculate_sentiment_score(comment):

    blob = TextBlob(comment)
    sentiment_score = blob.sentiment.polarity

    if sentiment_score > 0.5:
        return 5
    elif sentiment_score > 0.2:
        return 4
    elif sentiment_score > -0.2:
        return 3
    elif sentiment_score > -0.5:
        return 2
    else:
        return 1

data['SentimentRating'] = data['Full_Text'].apply(calculate_sentiment_score)
```

**Calculate Overall Rating from Comments**

```python
def calculate_overall_rating(sentiment_ratings):

    return round(sentiment_ratings.mean(),2)


sentiment_rating = calculate_overall_rating(data['SentimentRating'])
print("Overall Sentiment-Based Rating:", sentiment_rating)
```

**Product Prediction**

```python
# Step 4: Predict Good/Okay/Bad based on Sentiment Rating

def predict_rating_category(rating):
  if rating >= 4:
    return "Good"
  elif 2 <= rating < 4:
    return "Okay"
  else:
    return "Bad"

product_category = predict_rating_category(sentiment_rating)
print("Product Category (Sentiment-Based):", product_category)
```
